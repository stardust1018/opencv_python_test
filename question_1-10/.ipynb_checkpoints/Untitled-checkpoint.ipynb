```python
#
def gray(img):
    R = img[:,:,0]
    G = img[:,:,1]
    B = img[:,:,2]
    gray = R * 0.2126 + G * 0.7152 + B * 0.0722
    return np.clip(gray, 0, 255).astype(np.uint8)

img_bgr = cv2.imread("imori.jpg")
#  RGB
img_rgb = img_bgr[:,:,[2,1,0]]
img_gray = gray(img_rgb)
img_gray_std = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

#  1  2 
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# 
axes[0].imshow(img_rgb)
axes[0].set_title('img_rgb')
#axes[0].axis('off')  # 

axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('img_gray')
#axes[1].axis('off')

axes[2].imshow(img_gray_std, cmap='gray')
axes[2].set_title('img_gray_std')
#axes[2].axis('off')

plt.tight_layout()  # 
plt.show()
```

```python
#
def thre(img, t = 128):
    img_thre = img.copy()
    # NumPy
    img_thre[img_thre < t] = 0
    img_thre[img_thre >= t] = 255
    return img_thre

img_bgr = cv2.imread("imori.jpg")
#  RGB
img_rgb = img_bgr[:,:,[2,1,0]]
img_gray = gray(img_rgb)
img_thre = thre(img_gray)

#  1  2 
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# 
axes[0].imshow(img_rgb)
axes[0].set_title('img_rgb')
#axes[0].axis('off')  # 

axes[1].imshow(img_gray, cmap='gray')
axes[1].set_title('img_gray')
#axes[1].axis('off')

axes[2].imshow(img_thre, cmap='gray')
axes[2].set_title('img_thre')
#axes[2].axis('off')

plt.tight_layout()  # 
plt.show()
```

```python
# 
# 

#
def sfunc(img,t):
    w = img > t
    w1 = w.sum()/w.size
    w0 = 1-w1
    if w1==0 or w0==0:
        return 0
    m1 = np.mean(img[w])
    m0 = np.mean(img[~w])
    return w0*w1*(m0-m1)*(m0-m1)

#
def oturs(img):
    t = 0
    sd = 0
    for i in range(255):
        newsd = sfunc(img,i)
        if newsd > sd:
            sd = newsd
            t = i
    return t

img_bgr = cv2.imread("imori.jpg")
#  RGB
img_rgb = img_bgr[:,:,[2,1,0]]
img_gray = gray(img_rgb)
img_thre = thre(img_gray)
thread_oturs = oturs(img_gray)
img_oturs = thre(img_gray, thread_oturs)


plt.figure(figsize=(8, 8))
plt.subplot(2, 2, 1)
plt.imshow(img_rgb) # 
plt.title("img_rgb") # 
#plt.axis('off') # 

plt.subplot(2, 2, 2)
plt.imshow(img_gray, cmap='gray') # 
plt.title("img_gray") # 

plt.subplot(2, 2, 3)
plt.imshow(img_thre, cmap='gray')
plt.title("img_thre")

plt.subplot(2, 2, 4)
plt.imshow(img_oturs, cmap='gray')
plt.title("img_oturs")

plt.tight_layout()
plt.show()
```

```python
# rgb -- hsv

# HSVHueSaturationValue
def rgb2hsv(img):
    img = img.copy()
    #
    img = img / 255
    #img [128*128,3] 
    old_shape = img.shape
    img = img.reshape(-1,3)
    
    R,G,B = img[:,0],img[:,1],img[:,2]
    max = np.max(img, axis=1)
    min = np.min(img, axis=1)
    hsv = np.zeros_like(img)
    
    #
    h_b = min == B
    h_r = min == R
    h_g = min == G
    hsv[h_b,0] = 60*(G[h_b]-R[h_b])/(max[h_b]-min[h_b])+60
    hsv[h_r,0] = 60*(B[h_r]-G[h_r])/(max[h_r]-min[h_r])+180
    hsv[h_g,0] = 60*(R[h_g]-B[h_g])/(max[h_g]-min[h_g])+300
    #
    hsv[:,1] = max-min
    #
    hsv[:,2] = max
    
    return hsv.reshape(old_shape)

```
