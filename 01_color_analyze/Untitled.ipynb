```python
import cv2
print(cv2.__version__)

import numpy as np
print(np.__version__)

import matplotlib.pyplot as plt
```

```python
# 
img1 = cv2.imread("wide_101.png")
img2 = cv2.imread("tele_102.png")
# 
if img1 is None:
    raise FileNotFoundError("image1.jpg ")
if img2 is None:
    raise FileNotFoundError("image2.jpg ")

img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12, 6))  # 
# 
plt.subplot(1, 2, 1)  # 121
plt.imshow(img1_rgb)
plt.title("img1_rgb")
plt.axis("off")  # 
# 
plt.subplot(1, 2, 2)  # 122
plt.imshow(img2_rgb)
plt.title("img2_rgb")
plt.axis("off")  # 

plt.tight_layout()
plt.show()
```
