```python
import cv2 as cv
print(cv.__version__)
```

```python
img = cv.imread("open_logo.png")
# 如果图像路径错误或文件不存在，cv.imread() 会返回 None
if img is None:
    # 打印错误信息
    print("Error: Could not load image.")
    # 退出程序
    exit()
```

```python
cv.imshow("Display window", img)
cv.waitKey(0)
cv.destroyAllWindows()
```
