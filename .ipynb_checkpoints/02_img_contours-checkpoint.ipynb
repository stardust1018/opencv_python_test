```python
import cv2 as cv
print(cv.__version__)

import numpy as np
print(np.__version__)

import matplotlib.pyplot as plt
```

```python
img = cv.imread("bird_img.png")
# 如果图像路径错误或文件不存在，cv.imread() 会返回 None
if img is None:
    # 打印错误信息
    print("Error: Could not load image.")
    # 退出程序
    exit()
print(f"图像尺寸:height:{img.shape[0]}, width:{img.shape[1]}")
```

```python
def ShowImg(img, str):
    cv.imshow(str, img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
resize_img = cv.resize(img, (960, 720))
ShowImg(resize_img, "resize img")
print(f"resize_img图像尺寸:height:{resize_img.shape[0]}, resize_img:{resize_img.shape[1]}")
```

```python
# 转化为灰度图
gray_img = cv.cvtColor(resize_img, cv.COLOR_BGR2GRAY)
ShowImg(gray_img, "gray_img img")
```

```python
# 转化为二值图

# 简单阈值处理
ret, threshold_img = cv.threshold(gray_img, 127, 255, cv.THRESH_BINARY)
ShowImg(threshold_img, "threshold_img img")

# 自适应阈值处理
maxValue = 255
adaptiveMethod = cv.ADAPTIVE_THRESH_MEAN_C
thresholdType = cv.THRESH_BINARY
blockSize = 11
constNum = 2
adaptiveThreshold_img = cv.adaptiveThreshold(gray_img, maxValue, adaptiveMethod, thresholdType, blockSize, constNum)
ShowImg(adaptiveThreshold_img, "adaptiveThreshold_img img")


plt.figure(figsize=(15, 15))

plt.subplot(2, 2, 1)
threshold_rgb = cv.cvtColor(threshold_img, cv.COLOR_BGR2RGB)
plt.imshow(threshold_rgb) # 显示图像
plt.title("threshold_rgb") # 设置子图标题
plt.axis('off') # 关闭坐标轴

plt.subplot(2, 2, 2)
adaptiveThreshold_rgb = cv.cvtColor(adaptiveThreshold_img, cv.COLOR_BGR2RGB)
plt.imshow(adaptiveThreshold_rgb) # 显示图像
plt.title("adaptiveThreshold_rgb") # 设置子图标题
plt.axis('off') # 关闭坐标轴

plt.tight_layout()
plt.show()
```
